```python
import string
from collections import Counter
from nltk.corpus import stopwords

# Load the book text
with open("peter_pan.txt", encoding="utf-8") as f:
    text = f.read()

# Tokenize and preprocess
words = text.split()
words = [word.strip(string.punctuation).lower() for word in words]
words = [word for word in words if word.isalpha()]  # Remove numbers and punctuation
words = [word for word in words if word not in stopwords.words("english")]  # Remove stopwords

# Count word frequencies
word_counts = Counter(words)
most_common = word_counts.most_common(50)  # Top 50 to ensure we capture all characters

# List of likely character names (manually verified)
character_names = {"peter", "wendy", "john", "michael", "hook", "nana", "tinker", "tiger", "smee", "darling"}

# Extract top 10 most common meaningful words that are character names
protagonists = [word.capitalize() for word, count in most_common if word in character_names][:10]

# Final answer
protagonists
```
